```python
# bring interfaces up and either assign an address (if there is one) or flush address
from ipaddress import ip_address, IPv4Address, IPv4Network

for net in net_conf:
    for n in net['nodes']:
        if_name = n['name'] + '-' + net['name'] + '-p1'
        iface = slice.get_interface(if_name)
        iface.ip_link_up()
        if n['addr']:
            iface.ip_addr_add(addr=n['addr'], subnet=IPv4Network(net['subnet']))
        else:
            iface.get_node().execute("sudo ip addr flush dev %s"  % iface.get_device_name())
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define graph
G = nx.Graph()

# Define node types and shapes
node_types = {}
shapes = {'router': "o", 'client': "s", 'attacker': "h", 'server': "D"}  
# Router: circle, Client: square, Attacker: triangle, Server: diamond

# Retrieve all nodes in the slice and add them to the graph based on type
for node in slice.get_nodes():
    # Determine node type based on node name
    if "router" in node.get_name().lower():
        node_type = "router"
    elif "client" in node.get_name().lower():
        node_type = "client"
    elif "attacker" in node.get_name().lower():
        node_type = "attacker"
    elif "server" in node.get_name().lower():
        node_type = "server"
    else:
        continue  # Skip nodes that donâ€™t match our types
    
    # Add node to graph and set node type
    G.add_node(node.get_name())
    node_types[node.get_name()] = node_type

# Retrieve network connections and add edges between connected nodes
for network in slice.get_networks():
    # Get interfaces for each network connection
    interfaces = network.get_interfaces()
    if len(interfaces) == 2:  # Only consider point-to-point connections
        node1_name = interfaces[0].get_node().get_name()
        node2_name = interfaces[1].get_node().get_name()
        
        # Add edge between nodes without labels
        G.add_edge(node1_name, node2_name)

# Define colors based on node type
node_colors = []
for node in G.nodes():
    if node_types[node] == "router":
        node_colors.append("skyblue")
    elif node_types[node] == "client":
        node_colors.append("lightgreen")
    elif node_types[node] == "attacker":
        node_colors.append("salmon")
    elif node_types[node] == "server":
        node_colors.append("orange")

# Custom positions to center the server and unravel other nodes
pos = nx.spring_layout(G, seed=42, k=0.15, scale=3)
pos["Server"] = [0.5, 0.5]  # Fix the Server at the center
for node in pos:
    if node != "Server":  # Shift other nodes away from the center
        pos[node][0] += (pos[node][0] + 0.5) * 0.8
        pos[node][1] += (pos[node][1] + 0.5) * 0.8

# Draw the graph with different shapes
plt.figure(figsize=(14, 12))
plt.margins(0)  # Add padding around the plot

# Draw each type of node separately with its own shape
for node_type, shape in shapes.items():
    nx.draw_networkx_nodes(
        G, pos,
        nodelist=[node for node in G.nodes if node_types[node] == node_type],
        node_color=[node_colors[i] for i, node in enumerate(G.nodes) if node_types[node] == node_type],
        node_shape=shape,
        node_size=3000
    )

# Draw edges and add padding to labels
nx.draw_networkx_edges(G, pos, width=2)

# Apply padding to labels by offsetting them
label_pos = {node: (pos[node][0], pos[node][1] + 0.04) for node in pos}  # Offset labels slightly
nx.draw_networkx_labels(G, label_pos, font_size=10)

plt.title(f"Network Topology for Slice '{slice_name}' with Server Centered and Node Shapes")
plt.savefig("graph_paper.png", dpi=300)  # Set dpi to 300 or higher for better resolution
plt.show()

```

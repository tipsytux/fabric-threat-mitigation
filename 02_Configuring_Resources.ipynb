```python
# bring interfaces up and either assign an address (if there is one) or flush address
from ipaddress import ip_address, IPv4Address, IPv4Network

for net in net_conf:
    for n in net['nodes']:
        if_name = n['name'] + '-' + net['name'] + '-p1'
        iface = slice.get_interface(if_name)
        iface.ip_link_up()
        if n['addr']:
            iface.ip_addr_add(addr=n['addr'], subnet=IPv4Network(net['subnet']))
        else:
            iface.get_node().execute("sudo ip addr flush dev %s"  % iface.get_device_name())
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the network configuration
net_conf = [
    {"name": "net_server_firewall", "nodes": [("server", "firewall")]},
    {"name": "net_firewall_server_router", "nodes": [("firewall", "server_router")]},
    {"name": "net_server_router_isp1", "nodes": [("server_router", "isp_router1")]},
    {"name": "net_isp1_switch", "nodes": [("isp_router1", "switch")]},
    {"name": "net_isp2_switch", "nodes": [("isp_router2", "switch")]},
    {"name": "net_isp3_switch", "nodes": [("isp_router3", "switch")]},
    {"name": "net_isp2_client1", "nodes": [("isp_router2", "router_client1")]},
    {"name": "net_isp2_attacker2", "nodes": [("isp_router2", "router_attacker2")]},
    {"name": "net_isp3_client2", "nodes": [("isp_router3", "router_client2")]},
    {"name": "net_isp3_attacker1", "nodes": [("isp_router3", "router_attacker1")]},
    {"name": "net_isp3_attacker3", "nodes": [("isp_router3", "router_attacker3")]},
    {"name": "net_client1_router", "nodes": [("router_client1", "client1")]},
    {"name": "net_client2_router", "nodes": [("router_client2", "client2")]},
    {"name": "net_attacker1_router", "nodes": [("router_attacker1", "attacker1")]},
    {"name": "net_attacker2_router", "nodes": [("router_attacker2", "attacker2")]},
    {"name": "net_attacker3_router", "nodes": [("router_attacker3", "attacker3")]}
]

# Initialize a graph
G = nx.Graph()

# Add edges to the graph from the network configuration
for network in net_conf:
    G.add_edges_from(network["nodes"])

# Define custom shapes and colors for nodes
node_types = {
    "server": {"shape": "H", "color": "blue"},  # Hexagon
    "firewall": {"shape": "D", "color": "orange"},  # Diamond
    "router": {"shape": "o", "color": "green"},  # Oval
    "switch": {"shape": "s", "color": "pink"},  # Rectangle
    "client": {"shape": "^", "color": "yellow"},  # Square
    "attacker": {"shape": ">", "color": "red"}  # Square
}

# Helper function to determine node type
def get_node_type(node):
    if "router" in node:
        return "router"
    elif "firewall" in node:
        return "firewall"
    elif "server" in node:
        return "server"
    elif "switch" in node:
        return "switch"
    elif "client" in node:
        return "client"
    elif "attacker" in node:
        return "attacker"
    return "default"

# Function to draw the graph
def draw_network_graph(G, node_types):
    pos = nx.spring_layout(G, seed=42)  # Layout for better visualization
    plt.figure(figsize=(12, 8))

    # Draw nodes by type
    for node_type, style in node_types.items():
        nodes = [n for n in G.nodes if get_node_type(n) == node_type]
        nx.draw_networkx_nodes(
            G, pos,
            nodelist=nodes,
            node_shape=style["shape"],
            node_color=style["color"],
            node_size=700,
            edgecolors="black"
        )

    # Draw edges
    nx.draw_networkx_edges(G, pos, width=2, alpha=0.6, edge_color="gray")

    # Draw offset labels
    for node, (x, y) in pos.items():
        plt.text(
            x, y - 0.12,  # Offset slightly above the node
            node,
            fontsize=9,
            ha="center",
            va="center",
            bbox=dict(facecolor="white", alpha=0.5, edgecolor="black", boxstyle="round,pad=0.2")
        )

    plt.title("Simple Network Visualization", fontsize=14)
    plt.axis("off")
    plt.savefig("graph_paper.png", dpi=300)
    plt.show()

# Draw the graph with the custom shapes and colors
draw_network_graph(G, node_types)
```
